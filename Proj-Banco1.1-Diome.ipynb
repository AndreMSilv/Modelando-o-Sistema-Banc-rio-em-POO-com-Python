```python
SAUDACOES = """
  -=-=-=-=-=-=-=-=-=-=-=-=
    Bem vindo ao Banco Z
  -=-=-=-=-=-=-=-=-=-=-=-=
"""

CLIENTE = """
  -=-=-=-=-=-=-=-=-=-=-=-=
    Digite seu CPF para
      acessar a conta
  -=-=-=-=-=-=-=-=-=-=-=-=
"""


MENU = """
  -=-=-=-=-=-=-=-=-=-=-=-=
  Digite a opção desejada:
  [d] - Depósito
  [s] - Saque
  [e] - Extrato
  [f] - Fechar o programa
  -=-=-=-=-=-=-=-=-=-=-=-=
"""



LIMITE_VALOR_SAQUE = 500
LIMITE_SAQUES = 3
LIMITE_VALOR_TOTAL = LIMITE_VALOR_SAQUE * LIMITE_SAQUES
extrato = list()
saques = 0
valor_total_sacado = 0
saldo = 499
contas = [{"Nome":"Alexandre","Agencia":"0001","Cliente":12312312312,"Conta":"1","Saldo":saldo,"Extrato":extrato}]


print(SAUDACOES)


while True:

  print(CLIENTE)
  cpf_digitado = int(input("\n"))
  res = funcoes_projeto_2.chechar_cliente(cpf=cpf_digitado,contas=contas)
  print(res)

  print(MENU)
  opc = input("\n Digite a opção desejada: ")
  if opc in 'dDsSfFeE':
    
    # Verificando se usuário digitou numeros
    if opc in 'sS':
      valor = input('\nDigite o valor desejado para sacar: ')
      res = funcoes_projeto_2.saque(saldo=saldo,valor=valor,LIMITE_SAQUES=LIMITE_SAQUES,LIMITE_VALOR_SAQUE=LIMITE_VALOR_SAQUE)
      if res == False:
        continue
      else:
        extrato.append(['Saque',res])
        saldo -= res
        saques += 1
        continue

    # ESCOLHEU DEPOSITO
    
    elif opc in 'dD':
      # Verificando se usuário digitou numeros
      valor = input('\nDigite o valor desejado para depositar: ')
      try:
        valor = float(valor)
      except:
        print('\nSomente são aceitos valores monetários para depósito. Retornando ao menu principal')
        time.sleep(3)
        continue
  
      # Verificando condições do deposito
      if valor < 1:
        print("\nO valor mínimo para depósitos é de R$ 1.00 \nRetornando ao menu principal ")
        time.sleep(3)
        continue
      
      # Executando o depósito
      else:
        saldo += valor
        extrato.append(['Deposito',valor])
        print(f"\nVoce depositou R$ {valor:.2f} da sua conta, o saldo agora é de R$ {saldo:.2f}.")
        time.sleep(3)
        continue
          

    elif opc in 'eE':
      print("Imprimindo extrato: ")
      time.sleep(1)
      tamanho = len(extrato)
      for i in range(tamanho):
        print(f"Você realizou um {extrato[i][0]} no valor de R$ {extrato[i][1]:.2f} ")
        time.sleep(0.5)
        print(f"Seu saldo é de R$ {saldo:.2f}")
        time.sleep(5)

    else: 
      print('fechando')
      time.sleep(1)
      break
      
  else:
    print("\nOpção inválida, retornando ao menu principal \n")
    time.sleep(1)
```
